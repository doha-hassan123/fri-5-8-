```python
#Checking whether all values are defined in Euros or not
df.loc[df['value'].str.startswith('€')].shape[0]
```

```python
#Checking the same for Wages.
df.loc[df['wage'].str.startswith('€')].shape[0]
```

```python
#Splitting the value column to get just the numeric
df['value'] = df['value'].str.split('€')
df['value'] = df['value'].apply(lambda x:x[1])
```

```python
#Splitting the wage column to get just the numeric
df['wage'] = df['wage'].str.split('€')
df['wage'] = df['wage'].apply(lambda x:x[1])
```

```python
#Converting the player value in thousand Euros to Million Euros and then stripping the end denote.
df_value_K = df.loc[df['value'].str.endswith('K')]
df_value_K['value'] = df_value_K['value'].apply(lambda x: x[:-1])
df_value_K['value'] = df_value_K['value'].astype('float64')
df_value_K['value'] = df_value_K['value'] / 1000
```

```python
#Stripping the end denote for Million Euros Player value
df_value_M = df.loc[df['value'].str.endswith('M')]
df_value_M['value'] = df_value_M['value'].apply(lambda x: x[:-1])
df_value_M['value'] = df_value_M['value'].astype('float64')
```

```python
df_value_M.loc[df_value_M['wage'].str.endswith("M")]
```

```python
df2 = pd.concat([df_value_M,df_value_K])
```

```python
#Converting the player wage in thousand Euros to Million Euros and then stripping the end denote.
df_value_M['Wage'] =df_value_M['Wage'].apply(lambda x: x[:-1] if x.endswith('K') else x)
df_value_M['Wage'] = df_value_M['Wage'].astype('float64')
df_value_M['Wage'] = df_value_M['Wage'] / 1000
```
